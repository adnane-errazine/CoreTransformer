```python
import os
api_key = os.getenv("OPENAI_API_KEY")
print(api_key)
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {
            "role": "user",
            "content": "can you explain the meaning of 'cast' in the following safety guideline : Hateful content that also includes violence or serious harm towards the targeted group based on race, gender, ethnicity, religion, nationality, sexual orientation, disability status, or caste."
        }
    ]
)

print(completion.choices[0].message)
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a skillful artist."},
        {
            "role": "user",
            "content": "draw a picture of dorssaf"
        }
    ],
    
)

print(len(completion.choices))
for i in range(len(completion.choices)):
    print(completion.choices[i].message)
```

```python
response = client.images.generate(
  model="dall-e-3",
  prompt="north african woman sitting in a rural train station waiting for the train and holding her phone watching videos, and has a broken leg, she is about 1M64 and she weighs 77 Kg ",
  size="1024x1024",
  quality="hd",
  n=1,
)

image_url = response.data[0].url
```
